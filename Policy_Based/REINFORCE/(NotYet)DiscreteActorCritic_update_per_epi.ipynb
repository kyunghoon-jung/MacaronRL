```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torchsummary import summary

import gym
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import clear_output
from adabelief_pytorch import AdaBelief

# for using sampling with gradient-tracking when selecting an action
from torch.distributions import Categorical
```

```python
class Policy(nn.Module):

    def __init__(self, state_dim, action_dim, hidden):
        super(Policy, self).__init__()
        self.fc1 = nn.Linear(state_dim, hidden)
        self.fc2 = nn.Linear(hidden, action_dim)
        
    def forward(self, state):
        out = F.relu(self.fc1(state)) 
        out = F.softmax(self.fc2(out), dim=1) 
        return out 
```

```python
class Critic(nn.Module):
    
    def __init__(self, state_dim, action_dim, hidden):
        super(Critic, self).__init__()
        
        self.fc1 = nn.Linear(state_dim, hidden)
        self.fc2 = nn.Linear(hidden, 1) # one value per state
        
    def forward(self, state):
        out = F.relu(self.fc1(state))
        out = self.fc2(out)
        return out # Baseline value for REINFORCE with Baseline
```

```python
class Agent():
    def __init__(self, env, n_epi, max_steps, gamma, plot_freq, state_dim, action_dim, hidden, policy_learning_rate, critic_learning_rate, device):
        # environment parameter
        self.gamma = gamma
        self.env = env

        # trainig parameter
        self.device = device
        self.n_epi = n_epi
        self.max_steps = max_steps
        self.plot_freq = plot_freq
        self.frame_cnt = 0

        # network paramter
        self.p_lr = policy_learning_rate
        self.c_lr = critic_learning_rate
        self.policy = Policy(state_dim, action_dim, hidden).to(self.device)
        self.critic = Critic(state_dim, action_dim, hidden).to(self.device)
        self.policy_optimizer = optim.Adam(self.policy.parameters(), lr=self.p_lr)
        self.critic_optimizer = optim.Adam(self.critic.parameters(), lr=self.c_lr)
        
        # log parameter
        self.scores = []
        self.p_losses = []
        self.c_losses = []
        
    def select_action(self, state):
        prob = self.policy(torch.FloatTensor(state).view(1, -1).to(self.device)) 
        cate_dist = Categorical(prob) 
        action = cate_dist.sample() 
        return action.item(), cate_dist.log_prob(action), prob.detach().cpu().numpy() 
    
    def update_policy(self, epi_history):
        accum_R = 0
        p_loss = 0
        c_loss = 0
        
        for idx, [log_p, critic_value, delta] in enumerate(epi_history):
            p_loss += -log_p * delta
            c_loss += -critic_value * delta
        
        self.policy_optimizer.zero_grad()
        p_loss.backward()
        self.policy_optimizer.step()

        self.critic_optimizer.zero_grad()
        c_loss.backward()
        self.critic_optimizer.step()
        
        return [p_loss.item(), c_loss.item()] 
    
    def train(self):
        
        for i_episode in range(1, self.n_epi+1):
            epi_history = []
            rewards = [] 
            score = 0

            state = self.env.reset() 
            for step in range(self.max_steps):
                action, log_prob, prob = self.select_action(state)
                critic_value = self.critic(torch.FloatTensor(state).unsqueeze(0).to(device))
                
                next_state, reward, done, _ = env.step(action)
                target_value = reward + self.gamma*self.critic(torch.FloatTensor(next_state).unsqueeze(0).to(device))
                delta = (reward + (1-int(done))*self.gamma**target_value - critic_value).detach()
                
                epi_history.append([log_prob, critic_value, delta])
                state = next_state
                score += reward
                self.frame_cnt += 1
                if done: break
                    
            self.scores.append(score)
            loss = self.update_policy(epi_history)
            self.p_losses.append(loss[0]) 
            self.c_losses.append(loss[1])
            if i_episode%self.plot_freq == 0:
                self._plot_status(i_episode, self.p_losses, self.c_losses, self.scores)
    
    def _plot_status(self, i_episode, p_losses, c_losses, score_hist):
        subplot_params = [
            (311, f"Scores in episode_{i_episode}", score_hist),
            (312, f"Policy loss in episode:{i_episode}", p_losses),
            (313, f"Base loss in episode:{i_episode}", c_losses),
        ]

        clear_output(True)
        plt.figure(figsize=(20, 20), facecolor='w')
        for loc, title, values in subplot_params:
            plt.subplot(loc)
            plt.title(f'Frame:{self.frame_cnt} '+title)
            plt.plot(values)
        plt.show()
```

```python
device = torch.device(
            "cuda:1" if torch.cuda.is_available() else "cpu"
        )
env_name_list = ["CartPole-v0", "MountainCar-v0", "LunarLander-v2"]
env_name = env_name_list[0]
env = gym.make(env_name)
state_dim = env.observation_space.shape[0]
action_dim = env.action_space.n

print("Environment: ", env_name)
print("State Dimension:", state_dim, " Action Dimension:", action_dim)

n_epi = 100000 
max_steps = 1000
gamma = 0.99
plot_freq = 10

hidden = 64
policy_learning_rate = 0.0001
critic_learning_rate = 0.001
```

```python
agent = Agent(
            env,
            n_epi,
            max_steps,
            gamma,
            plot_freq,
            state_dim,
            action_dim,
            hidden,
            policy_learning_rate,
            critic_learning_rate,
            device)
```

```python
agent.train() 
```
